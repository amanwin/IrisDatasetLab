# Dataset

Iris plants dataset is a benchmark dataset for testing supervised machine learning problems. It comprises of four attributes (sepal length, sepal width, petal length, petal width) and 50 samples from each of the three different species (classes) of the iris flower. 








```python
#Import libraries

from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

1. Load the dataset from the sklearn library. **(1 point)**
2. Explore the dataset by printing attribute names, attribute values, and label names and values. **(4 points)** {1*4=4}

```python
# Load the iris dataset together with the target labels
iris = load_iris()
print("***************************************************************************************")
print("Attribute names:\n")
print(iris.feature_names)
print("***************************************************************************************")
print("Attribute values:\n")
print(iris['data'])
print("Iris data shape:\n")
print(iris['data'].shape)
print("***************************************************************************************")
print("Target:\n")
print(iris['target'])
print("Target shape:\n")
print(iris['target'].shape)
print("***************************************************************************************")
print("Label names:\n")
print(iris['target_names'])
```

```python
X = iris['data']
y = iris['target']
```

```python
iris_df = pd.DataFrame(X, columns=iris['feature_names'])
iris_target_df = pd.DataFrame(X, columns=iris['feature_names'])
iris_target_df['class'] = iris['target']
iris_df.head()
```

Plot a histogram indicating the distribution of different attributes in the dataset. {Expected output is four histograms, each corresponding to a different attribute. The X-axis is the domain of the attribute, and Y-axis is the frequency of the attribute corresponding to a domain}        **(4 points)** {1*4=4}

```python
def plot_histogram(col,title, xlabel,color="orange"):
    plt.figure(figsize=(10,7))
    x = col
    plt.hist(x, bins=20,color = color, lw=0)
    plt.title(title) 
    plt.xlabel(xlabel) 
    plt.ylabel("Count")
    plt.show()
    
```

```python
plot_histogram(iris_df['sepal length (cm)'], 'Sepal Length in cm','Sepal_Length_cm')
plot_histogram(iris_df['sepal width (cm)'], 'Sepal Width in cm','Sepal_Width_cm')
plot_histogram(iris_df['petal length (cm)'], 'Petal Length in cm','Petal_Length_cm')
plot_histogram(iris_df['petal width (cm)'], 'Petal Width in cm','Petal_Width_cm')
```


“Data Standardization” means that the range of values is “standardized to measure how many standard deviations the value is from its mean”. 
1. Standardize the data and print the standardized data in a tabular form. **(1 point)** 
2. Plot the histogram corresponding to each attribute as done above. **(4 points)** {1*4=4}

```python

# Standardize the dataset 
standardized_iris_df = preprocessing.scale(iris_df)
standardized_iris_df = pd.DataFrame(standardized_iris_df, columns=iris['feature_names'])


# Print dataset before standardization        
print("Dataset before standardization:\n")
print(iris_df)

# Print dataset after standardization  
print("Dataset after standardization:\n")
print(standardized_iris_df)

# For entire standardized dataset, plot 4 histograms for sepal length (cm), sepal width (cm), petal length (cm) and petal width (cm)
plot_histogram(standardized_iris_df['sepal length (cm)'], 'Sepal Length in cm','Sepal_Length_cm')
plot_histogram(standardized_iris_df['sepal width (cm)'], 'Sepal Width in cm','Sepal_Width_cm')
plot_histogram(standardized_iris_df['petal length (cm)'], 'Petal Length in cm','Petal_Length_cm')
plot_histogram(standardized_iris_df['petal width (cm)'], 'Petal Width in cm','Petal_Width_cm')
```

### Note
Standardization refers to shifting the distribution of each attribute to have a mean of zero and a standard deviation of one (unit variance). The same can be depicted in the histogram.

Let's study the relationship between attributes.
1. Find Pearson correlation for each pair of attributes. {Expected output is a table (triangular matrix)} **(2 points)** 
2. Also, draw a heatmap representing this correlation. **(2 points)** 

```python
# Compute Pearson correlation between each pair of attributes
corr_matrix = iris_df[['sepal length (cm)','sepal width (cm)', 'petal length (cm)', 'petal width (cm)']].corr()

# The table is going to be symmetric, so you can make a triangular matrix/table 
corr_matrix

# Generate a heatmap according to the person coefficient values 
ax = sns.heatmap(corr_matrix, annot = True, cmap="Set1")
top, bottom = ax.get_ylim()
ax.set_ylim(top+0.5, bottom-0.5)
```
